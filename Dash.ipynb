```python
df = pd.read_csv('/content/spotify-music-dataset/high_popularity_spotify_data.csv')
```

```python
from dash import Dash, dcc, html
import plotly.express as px
from IPython.display import display, HTML
import pandas as pd

# Пример DataFrame
data = df.copy()  # твой датасет

# Обработка даты
data['track_album_release_date'] = pd.to_datetime(
    data['track_album_release_date'], errors='coerce'
)
data['year'] = data['track_album_release_date'].dt.year

# Графики
fig_genre = px.box(
    data,
    x='playlist_genre',
    y='track_popularity',
    title='Популярность треков по жанрам'
)

fig_energy = px.scatter(
    data,
    x='energy',
    y='track_popularity',
    opacity=0.5,
    title='Energy vs Popularity'
)

energy_by_year = data.groupby('year')['energy'].mean().reset_index()
fig_trend = px.line(
    energy_by_year,
    x='year',
    y='energy',
    title='Изменение Energy по годам'
)

mean_popularity = round(data['track_popularity'].mean(), 1)

# Создание Dash приложения
app = Dash(__name__)

app.layout = html.Div(
    style={'padding': '20px'},
    children=[
        html.H1("Музыкальные треки и популярность"),
        html.H3(f"Средняя популярность треков: {mean_popularity}"),
        dcc.Graph(figure=fig_genre),
        dcc.Graph(figure=fig_energy),
        dcc.Graph(figure=fig_trend)
    ]
)

# Запуск сервера на localhost с внешней ссылкой для Jupyter
# Выведет ссылку для браузера
app.run_server(debug=True, port=8050, mode='external')
```
